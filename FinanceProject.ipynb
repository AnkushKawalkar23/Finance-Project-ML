```python
# Define mapping for buckets
def assign_bucket(age):
    if age <= 24:
        return '24 yrs and below'
    elif 25 <= age <=34:
        return '25 yrs to 34 yrs'
    elif 35 <= age <=44:
        return '35 yrs to 44 yrs'
    else:
        return '45 yrs & Above'
```

```python
# Check if existing bucket matches expected one
df1['Expected_Bucket'] = df1['Age'].apply(assign_bucket)
mismatch = df1[df1['Age_Bucket'] !=  df1['Expected_Bucket']]

if mismatch.empty:
    print('All age value correctly match')
else:
    print("Mismatches found â€“ investigate before dropping Age column")
    display(mismatches[["Age", "Age_Bucket", "Expected_Bucket"]])
```

```python
def get_device_brand(device):
    if pd.isna(device):
        return 'Unknown'
    device = device.lower()
    if 'iphone' in device or 'ipad' in device:
        return 'Apple'
    elif "samsung" in device:
        return 'Samsung'
    elif 'oneplus' in device:
        return "One-plus"
    elif 'redmi' in device or 'xiaomi' in device or 'mi' in device:
        return "Xiaomi/Redmi"
    else:
        return "Others"
```

```python
# apply to column 
df1['Device_Brand'] = df1['device_model'].apply(get_device_brand)

#check of counts
print(df1["Device_Brand"].value_counts())
```

```python
# mapping for city 
def get_city_name(city):
    if pd.isna(city):
        return 'Unknown'
    elif ("Mumbai" in city or 'Virar' in city or 'Dombivli' in city or 'New Panvel' in city 
          or 'Panvel' in city or 'Kalyan' in city or 'Vikhroli' in city or 'Ghatkopar' in city
          or 'Navi Mumbai Panvel' in city or 'Panchgani' in city or 'Bhiwandi' in city or 'Navi Mumbai' in city):
        return 'Mumbai'
    elif ('Bengaluru' in city or 'Bangalore Urban' in city or 'bangalore' in city or 'Bangalore Rural' in city):
        return 'Bengaluru'
    elif ('Delhi NCR' in city or 'Delhi' in city or 'East Delhi' in city or 'West Delhi' in city 
          or 'Gautam Buddha Nagar' in city or 'Noida' in city or 'New Delhi' in city or 'North West Delhi' in city
          or 'Central Delhi' in city or 'North Delhi' in city or 'Greater Noida' in city 
          or 'South Delhi' in city or 'Pitampura' in city or 'North East Delhi' in city or 'Karol Bagh' in city or 'Gurgaon' in city or 'Noida' in city
         or "Rohini" in city or 'Greater Noida' in city or 'Ghaziabad' in city):
        return 'Delhi/ Delhi-NCR'
    elif ('Pune' in city or 'Hadapsar, Pune' in city or 'Pune Division' in city):
        return 'Pune'
    elif 'Kolkata' in city:
        return 'Kolkata'
    elif 'Chennai' in city:
        return 'Chennai'
    elif 'Hyderabad' in city:
        return 'Hyderabad'
    elif 'Ahmedabad' in city:
        return 'Ahmedabad'
    else:
        return 'Tier II & below'
```

```python
# apply to column
df1['City_Updated'] = df1['City'].apply(get_city_name)

```
